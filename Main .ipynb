```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
data = pd.read_csv("Iris.csv")
```

```python
data.head(50)
```

```python
data.columns
```

```python
X = data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']] 
y = data['Species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print(f"Training set size: {X_train.shape}")
print(f"Testing set size: {X_test.shape}")
```

```python
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")
```

```python
plt.figure(figsize=(15, 10))
plot_tree(clf, feature_names=X.columns, class_names=clf.classes_, filled=True)
plt.title("Decision Tree Visualization")
plt.show()
```

```python
import streamlit as st
import joblib
import numpy as np

# Load the saved model
model = joblib.load('decision_tree_model.pkl')

# Title of the web app
st.title('Iris Species Prediction with Decision Tree')

# Input form for user to enter flower measurements
SepalLengthCm = st.number_input('Sepal Length (cm)', min_value=0.0)
SepalWidthCm = st.number_input('Sepal Width (cm)', min_value=0.0)
PetalLengthCm = st.number_input('Petal Length (cm)', min_value=0.0)
PetalWidthCm = st.number_input('Petal Width (cm)', min_value=0.0)

# When the user clicks "Predict"
if st.button('Predict'):
    # Prepare the input data
    features = np.array([[
        SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm
    ]])

    # Make prediction
    prediction = model.predict(features)

    # Show the result
    st.write(f'The predicted species is: {prediction[0]}')
```
